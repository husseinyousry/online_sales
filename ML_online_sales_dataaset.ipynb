### droping the unwanted columns

```python
df_copy.drop(['invoice_num', 'stock_code', 'description', 'quantity',
       'customer_id', 'country',  
       'category', 'sales_channel', 'return_status', 'shipment_provider',
       'warehouse_location',  'order_year', 'order_month',
       'order_weekday', 'order_hour', 'is_return', 'net_quantity',
       'sales'], axis=1, inplace=True)
```

### handling categorical columns using label encoder

```python
encoder = {
    'Low': 1,
    'Medium': 2,
    'High': 3
}
df_copy['order_priority'] = df_copy['order_priority'].map(encoder)
```

```python
encoder_1 = {False: 0, True: 1}
df_copy['is_refund'] = df_copy['is_refund'].map(encoder_1)
```

```python
df_copy.head(2)
```

### handling cetegorical data using onehotencoder

```python
df_copy = pd.get_dummies(df_copy,columns=['payment_method'], drop_first=True,dtype='int')
df_copy.head(2)
```

### split the input and the output

```python
x = df_copy.drop(['is_refund'], axis=1)
y = df_copy['is_refund']
```

### split the data to train and test

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

### scaling

```python
scaler = StandardScaler()
x_scaled_train = scaler.fit_transform(x_train)
x_scaled_test = scaler.transform(x_test)
```

```python
x_scaled_train = pd.DataFrame(x_scaled_train, columns=x_train.columns)
x_scaled_test = pd.DataFrame(x_scaled_test, columns=x_test.columns)
```

```python
smote = SMOTE()
x_train_resampled, y_train_resampled = smote.fit_resample(x_scaled_train, y_train)
```

```python
x_train_resampled = pd.DataFrame(x_train_resampled, columns=x_scaled_train.columns)
x_train_resampled.head(2)
```
